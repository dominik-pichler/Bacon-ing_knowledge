```python
import sqlite3
import pandas as pd
```

```python
import spacy

# Load the German transformer model
nlp = spacy.load("de_dep_news_trf")

def tokenize_and_extract_entities(text):
    # Process the text with the loaded model
    doc = nlp(text)

    # Tokenize: Print each token in the text
    print("Tokens:")
    for token in doc:
        print(f"Token: {token.text}, POS: {token.pos_}, Lemma: {token.lemma_}")

    # Extract entities
    entities = []
    print("\nEntities:")
    for ent in doc.ents:
        print(f"Entity: {ent.text}, Label: {ent.label_}")
        entities.append({
                "text": ent.text,
                "start": ent.start_char,
                "end": ent.end_char,
                "label": ent.label_
            })
    
    return entities

# Example usage
text = "ich wohne in Westfalen"
geo_entities = tokenize_and_extract_entities(text)
print("\nGeographical Entities:", geo_entities)
```

```python
def run_query(database_path, query):
    # Connect to the SQLite database
    # If the database does not exist, it will be created
    connection = sqlite3.connect(database_path)
    
    try:
        # Create a cursor object using the connection
        cursor = connection.cursor()
        
        # Execute the SQL query
        cursor.execute(query)
        
        # Fetch all results from the executed query
        results = cursor.fetchall()
        
        # Turning list of tuples to list of strings for further processing
        return [str(t[0]) for t in results]

            
    except sqlite3.Error as e:
        print(f"An error occurred: {e}")
        
    finally:
        
        # Close the connection to the database
        if connection:
            connection.close()
```

```python
database_path = 'openregister.db'  # Path to your SQLite database file
query = 'SELECT name FROM company LIMIT 100;'  # Replace 'your_table' with your actual table name

company_names = run_query(database_path, query)
company_names
```
